```python
ph = float(input("Di el valor del ph"))

if ph == 7 :
  print("El ph es neutro")
elif ph <= 4 :
  print ("El ph es poco acido")
elif ph == 0 :
  print ("El ph es muy acido")
elif ph >7 and ph <9:
  print ("El ph es poco alcalino")
elif ph >=10 and ph<=14 :
  print("El ph es muy alcalino")
```
